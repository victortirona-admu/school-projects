```python
#Importing all libraries
import nltk
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
%pprint
```

```python
#importing text files into lists
networks = open("data/networks.txt").read()
networks = networks.split(" ")
stopwords = open("data/stop.txt").read()
stopwords = stopwords.replace("'","")
stopwords = stopwords.replace(",", "")
stopwords = stopwords.replace('"',"")
stopwords = stopwords.split("\n")
print(stopwords)
```

```python
#Creating a dictionary for networks.txt
dict_networks = {} #dict_networks[key] = values

for word in networks:
    ctr=0
    for everyword in networks:
        if (word == everyword):
            ctr = ctr +1
    dict_networks[word] = ctr
    
#Plotting networks.txt before the stopwords are removed
height = dict_networks.values()
bars = dict_networks.keys()
y_pos = np.arange(len(bars)) # count to the number of words
 
plt.figure(figsize=(10,100))
    
plt.barh(y_pos, height, 5)
 
plt.yticks(y_pos, bars)

plt.show()
```

```python
#removing stopwords
filtered_networks = []
filtered_networks = [word for word in networks if word not in stopwords]
filtered_networks

#creating a new dictionary
dict_filt_networks = {}

for word in filtered_networks:
    ctr=0
    for everyword in filtered_networks:
        if (word == everyword):
            ctr = ctr +1
    dict_filt_networks[word] = ctr
    
#plotting filtered networks dictionary
height = dict_filt_networks.values()
bars = dict_filt_networks.keys()
y_pos = np.arange(len(bars)) # count to the number of words
 
plt.figure(figsize=(10,100))
    
plt.barh(y_pos, height, 5)
 
plt.yticks(y_pos, bars)

plt.show()
```

### Function for plurality

```python
#This function checks the last 4 letters of each english word and compares it to common plural suffixes. If it matches, it adds a "mga" to the Filipino word
#this is found inside the creatEntoFilDictionary function
def addMga(word, i):
    #initialize list of plural suffixes
    oneletter = ["s"]
    twoletter = ["es"]
    threeletter = ["ses","xes","zes","ies","ves"]
    fourletter = ["sses","shes","ches"]
    #initializing the variables for the last characters of a letter
    #they are initialized as commas because they need to not be found in the lists above for if else validity reasons. In case an if else statement calls them, the system will not error!
    last4=","
    last3=","
    last2=","
    
    #Checks the length of the word and stores last four characters (plural suffixes are only up to 4 characters long)
    if (len(word)>3):
        last4 = word[-4] + word[-3] + word[-2] + word[-1]
    if (len(word)>2):
        last3 = word[-3] + word[-2] + word[-1]
    if (len(word)>1):
        last2 = word[-2] + word[-1]
    last1 = word[-1]
                
    #compares the stores last four characters to the lists of plural suffixes. This starts with four-letter plural suffixes and works their way down to one-letter plural suffix because one-letter suffixes are present in two-letter suffixes and so on
    if last4 in fourletter:
        translator[i-1] = "mga " + translator[i-1] #adds a "mga" in the filipino noun because it is plural!
    elif last3 in threeletter:
        translator[i-1] = "mga " + translator[i-1] 
    elif last2 in twoletter:
        translator[i-1] = "mga " + translator[i-1] 
    elif last1 in oneletter:
        translator[i-1] = "mga " + translator[i-1]
            
#testing   
#i=1
#translator = ["bubong", "puppies"]
#translator = open("data/tl-en.txt", encoding="utf8").read()
#translator = translator.replace("\t","\n") #removing those weird arrow things (tabs)
#translator = translator.split("\n") #every even number in the list is tagalog and every odd number is english
#addMga(translator[i])
```

### Function for multiple entries

```python
#This function handles multiple entries by concatenating them together as one dictionary entry separated by a "/"
#Found inside the main code for problem 4
def createEntoFilDictionary(translator):
    for i in range(1,len(translator),2): #increments by 2 because we look at every PAIR of words in the list
        addMga(translator[i],i) #plurality
        if translator [i] in dict_tlen.keys(): #if key is already in dictionary,
            dict_tlen[translator[i]]=dict_tlen[translator[i]] + "/" + translator[i-1] #if english word is already in the dictionary, it'll just concatenate the other translations with the existing translation
            #print(translator[i] + " is already in dictionary, will now add " + translator [i-1] )
        else:
            dict_tlen[translator[i]]=translator[i-1] # "i" is english word and "i-1" is the filipino word
```
